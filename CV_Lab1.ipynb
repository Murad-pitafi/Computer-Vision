```python
# 1. Grocery List: Your task is to create a simple program to manage a grocery list. Your program should allow the user to add
# items to the list, remove items, and display the current list of items. Write a Python program that implements this grocery list
# management system.

def add_item(grocery_list, item):
    grocery_list.append(item)
    print(f"'{item}' added to the grocery list.")

def remove_item(grocery_list, item):
    if item in grocery_list:
        grocery_list.remove(item)
        print(f"'{item}' removed from the grocery list.")
    else:
        print(f"'{item}' not found in the grocery list.")

def display_list(grocery_list):
    if grocery_list:
        print("Current Grocery List:")
        for i, item in enumerate(grocery_list, start=1):
            print(f"{i}. {item}")
    else:
        print("The grocery list is empty.")

def main():
    grocery_list = []

    while True:
        print("\nGrocery List Management")
        print("1. Add Item")
        print("2. Remove Item")
        print("3. Display List")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            item = input("Enter the item to add: ")
            add_item(grocery_list, item)

        elif choice == '2':
            item = input("Enter the item to remove: ")
            remove_item(grocery_list, item)

        elif choice == '3':
            display_list(grocery_list)

        elif choice == '4':
            print("Exiting Grocery List Management System.")
            break

        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```

```python
# 2. Student Record System: Your task is to build a simple student record system. Each student is identified by their student ID,
# and you need to store their names and corresponding grades. Your program should allow adding students, updating their
# grades, and displaying the student records. Write a Python program that implements this student record system using a
# dictionary.


def add_student(records, student_id, name, grade):
    if student_id in records:
        print(f"Student ID {student_id} already exists. Use update option to change the grade.")
    else:
        records[student_id] = {'Name': name, 'Grade': grade}
        print(f"Student {name} added successfully.")

def update_grade(records, student_id, grade):
    if student_id in records:
        records[student_id]['Grade'] = grade
        print(f"Grade updated for student ID {student_id}.")
    else:
        print(f"Student ID {student_id} not found. Add the student first.")

def display_records(records):
    if records:
        print("Student Records:")
        for student_id, details in records.items():
            print(f"ID: {student_id}, Name: {details['Name']}, Grade: {details['Grade']}")
    else:
        print("No records found.")

def main():
    student_records = {}

    while True:
        print("\nStudent Record System")
        print("1. Add Student")
        print("2. Update Grade")
        print("3. Display Records")
        print("4. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            student_id = input("Enter student ID: ")
            name = input("Enter student name: ")
            grade = input("Enter student grade: ")
            add_student(student_records, student_id, name, grade)

        elif choice == '2':
            student_id = input("Enter student ID to update: ")
            grade = input("Enter new grade: ")
            update_grade(student_records, student_id, grade)

        elif choice == '3':
            display_records(student_records)

        elif choice == '4':
            print("Exiting Student Record System.")
            break

        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```

```python
# 3. Load an image from file and display it, convert to grayscale, resize it to specific size using OpenCV.

import cv2  
image = cv2.imread('image.jpg')

 
cv2.imshow('Original Image', image)
cv2.waitKey(0)

 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale Image', gray_image)
cv2.waitKey(0)

 
resized_image = cv2.resize(image, (300, 300))
cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 4. Create a blank image and draw basic shapes like rectangles and circles on it using OpenCV.


import numpy as np
 
blank_image = np.zeros((400, 400, 3), dtype='uint8')
 
cv2.rectangle(blank_image, (50, 50), (300, 300), (255, 0, 0), 3)

cv2.circle(blank_image, (200, 200), 50, (0, 255, 0), -1)
 
cv2.imshow('Shapes', blank_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 5. Load an image and apply Gaussian blur, crop at specific region using OpenCV, NumPy array slicing.


 
 
image = cv2.imread('image.jpg')

 
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)
cv2.imshow('Gaussian Blurred Image', blurred_image)
cv2.waitKey(0)

cropped_image = image[100:300, 100:300]
cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 7. Load a grayscale image and apply binary thresholding to it and the rotate at 60 0 using OpenCV.



 
image = cv2.imread('image.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 
_, thresh_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
cv2.imshow('Thresholded Image', thresh_image)
cv2.waitKey(0)
 
(h, w) = thresh_image.shape[:2]
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, 60, 1.0)
rotated_image = cv2.warpAffine(thresh_image, matrix, (w, h))
cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 8. Load two images and blend them using OpenCV, and the convert it to grayscale and apply histogram equalization to enhance
# contrast.
 
image1 = cv2.imread('image.jpg')
image2 = cv2.imread('image_2.jpg')
 
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
 
blended_image = cv2.addWeighted(image1, 0.5, image2, 0.7, 0)
cv2.imshow('Blended Image', blended_image)
cv2.waitKey(0)
 
gray_blended_image = cv2.cvtColor(blended_image, cv2.COLOR_BGR2GRAY)
 
equalized_image = cv2.equalizeHist(gray_blended_image)
cv2.imshow('Equalized Image', equalized_image)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# 9. Create binary images and perform bitwise AND, OR, XOR, and NOT operations using OpenCV and NumPy.
 
image1 = np.zeros((400, 400), dtype='uint8')
image2 = np.zeros((400, 400), dtype='uint8')
 
cv2.rectangle(image1, (100, 100), (300, 300), 255, -1)
cv2.circle(image2, (200, 200), 100, 255, -1)
 
bitwise_and = cv2.bitwise_and(image1, image2)
bitwise_or = cv2.bitwise_or(image1, image2)
bitwise_xor = cv2.bitwise_xor(image1, image2)
bitwise_not = cv2.bitwise_not(image1)
 
cv2.imshow('Bitwise AND', bitwise_and)
cv2.imshow('Bitwise OR', bitwise_or)
cv2.imshow('Bitwise XOR', bitwise_xor)
cv2.imshow('Bitwise NOT', bitwise_not)
cv2.waitKey(0)

cv2.destroyAllWindows()
```
